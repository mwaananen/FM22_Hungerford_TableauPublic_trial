```python
import pandas as pd
```

```python
#Football Manager 22 Personal Save - Hungerford Town F.C.
fm22 = "MoneyBall Experiment 1.csv"
```

```python
hungerford_fm22_all_df = pd.read_csv(fm22, encoding="UTF-8")
```

```python
hungerford_fm22_all_df.head()
```

```python
#Must Remove all punctuation from columns
#Including 
# - , $ / %
#And change data types in several numeric columns

old_value = '-'
new_value = '0'
hungerford_fm22_all_df.replace(old_value, new_value, inplace=True)
```

```python
hungerford_fm22_all_df.head()
```

```python
#Remove p/a (per annual) and the /$, puntucation from Salary Column
hungerford_fm22_all_df["Salary"] = hungerford_fm22_all_df["Salary"].replace('p/a', "", regex=True)

#Change to float
hungerford_fm22_all_df['Salary'] = hungerford_fm22_all_df['Salary'].replace('[/$,]',"", regex=True).astype(float)
```

```python
#Remove the $ from Asking Price column
hungerford_fm22_all_df['Asking Price'] = hungerford_fm22_all_df['Asking Price'].replace('[$]',"", regex=True)

#Change to float and convert K and M to numeric thousand and million, respectfully, and c
hungerford_fm22_all_df['Asking Price'] = hungerford_fm22_all_df['Asking Price'].replace({"K":"*1e3", "M":"*1e6"}, regex=True).map(pd.eval).astype(float)

#Check new dtype
hungerford_fm22_all_df.dtypes[['Salary', 'Asking Price']]
```

```python
#Remove the % from all relevant columns
hungerford_fm22_all_df[['Shot %', 'Hdr %', 'Pas %', 'xSv %', 'Sv %', 'Pen/R', 'Cr C/A', 'Tck R', 'Pens Saved Ratio']] = hungerford_fm22_all_df[['Shot %', 'Hdr %', 'Pas %', 'xSv %', 'Sv %', 'Pen/R', 'Cr C/A', 'Tck R', 'Pens Saved Ratio']].replace('%',"", regex=True).astype(float)

#check data types
hungerford_fm22_all_df.dtypes[['Shot %', 'Hdr %', 'Pas %', 'xSv %', 'Sv %', 'Pen/R', 'Cr C/A', 'Tck R', 'Pens Saved Ratio']]
```

```python
#Remove the , from all total minutes column
hungerford_fm22_all_df[['Mins', 'Pas A', 'Ps C']] = hungerford_fm22_all_df[['Mins', 'Pas A', 'Ps C']].replace(',',"", regex=True).astype(float)

#check data types
hungerford_fm22_all_df.dtypes[['Mins', 'Pas A', 'Ps C']]
```

```python
#remove "lbs", change data type, and convert from pounds to kilograms
hungerford_fm22_all_df['Weight'] = hungerford_fm22_all_df['Weight'].replace('lbs', "", regex=True).astype(float)
hungerford_fm22_all_df['Weight'] = hungerford_fm22_all_df['Weight'] * 0.454 #convert lbs to kgs

#in original excel file chnage height in feet/inches to total inches in new excel column
#new column is labeled as the "Inches" column
#change data type, and convert from inches to centimeters
hungerford_fm22_all_df['Inches'] = hungerford_fm22_all_df['Inches'].astype(float)
hungerford_fm22_all_df['Inches'] = hungerford_fm22_all_df['Inches'] * 2.54 #convert inches to cms

#check data types
hungerford_fm22_all_df.dtypes[['Weight', 'Inches']]
```

```python
#rename columns to represent new measurement
hungerford_fm22_all_df.rename(columns={"Inches": "Centimeters", "Weight": "Kilograms"}, inplace=True)
```

```python
#check inch to centimeter and pounds to kilogram conversions
hungerford_fm22_all_df.loc[:,["Centimeters","Kilograms"]]
```

```python
#remove feet/inches column that was replaced  with Inches column
hungerford_fm22_all_df.drop(['Height'], axis=1, inplace=True)
```

```python
#change all columns from object to float that represent numeric measurements,
#and that have not had punctuation cleaned out of column in above cells
#NOTE 'Hdrs' may need to have punctuation (,) cleaned by end of season totals
hungerford_fm22_all_df[[
    'Age',
    'Starts',
    'Mins/Gm',
    'Av Rat',
    'Gls',
    'Gls/90',
    'Shots',
    'ShT',
    'Shot/90',
    'ShT/90',
    'Pens',
    'Pens S',
    'Off',
    'xG',
    'CCC',
    'Ch C/90',
    'Ast',
    'Asts/90',
    'K Pas',
    'K Ps/90',
    'Ps A/90',
    'Ps C/90',
    'Cr A',
    'Cr C',
    'Drb',
    'Drb/90',
    'FA',
    'Hdrs A',
    'Aer A/90',
    'Hdrs',
    'Hdrs W/90',
    'K Tck',
    'Tck W',
    'Tck',
    'Fls',
    'Yel',
    'Red',
    'Itc',
    'Int/90',
    'Clear',
    'Conc',
    'All/90',
    'Shutouts',
    'Svh',
    'Svt',
    'Svp',
    'Pens Faced',
    'Pens Saved',
    'PoM'
]] = hungerford_fm22_all_df[[
    'Age',
    'Starts',
    'Mins/Gm',
    'Av Rat',
    'Gls',
    'Gls/90',
    'Shots',
    'ShT',
    'Shot/90',
    'ShT/90',
    'Pens',
    'Pens S',
    'Off',
    'xG',
    'CCC',
    'Ch C/90',
    'Ast',
    'Asts/90',
    'K Pas',
    'K Ps/90',
    'Ps A/90',
    'Ps C/90',
    'Cr A',
    'Cr C',
    'Drb',
    'Drb/90',
    'FA',
    'Hdrs A',
    'Aer A/90',
    'Hdrs',
    'Hdrs W/90',
    'K Tck',
    'Tck W',
    'Tck',
    'Fls',
    'Yel',
    'Red',
    'Itc',
    'Int/90',
    'Clear',
    'Conc',
    'All/90',
    'Shutouts',
    'Svh',
    'Svt',
    'Svp',
    'Pens Faced',
    'Pens Saved',
    'PoM'
]].astype(float)
```
